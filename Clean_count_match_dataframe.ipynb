```python
list_title = []
list_lyrics = []
list_artis = []
for filename in os.listdir(r"C:\Users\lea_r\Desktop\NLP\json_lyrics"):
    with open(os.path.join(r"C:\Users\lea_r\Desktop\NLP\json_lyrics", filename), 'r') as json_file:
        data = json.load(json_file)        
        list_title.append(data["title"])
        list_lyrics.append(data["lyrics"])
        list_artis.append(data["artist"])   
        with open ("data_lyrics.txt", "w") as f:
            json.dump(list_lyrics, f)
      
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
with open ("data_lyrics.txt", "r") as file:
    text = file.read()
```

```python
werbung = "EmbedShare URLCopyEmbedCopy"
separator = "\\n"
chorus = "[Chorus]"
bridge = "[Bridge]"
outro = "[Outro]"
interlude = "[Interlude]"
verse = "[Verse ]"
verse2 = "[Verse]"
intro = "[Intro]"
postChorus = "[Post-Chorus]"
preChorus = "[Pre-Chorus]"
hook = "[Hook ]"


text = text.replace(werbung, "")
text = text.replace(separator, " ")
text = text.replace(chorus, "")
text = text.replace(bridge, "")
text = text.replace(outro, "")
text = text.replace(interlude, "")
text = text.replace(verse, "")
text = text.replace(verse2, "")
text = text.replace(intro, "")
text = text.replace(postChorus, "")
text = text.replace(preChorus, "")
text = text.replace(hook, "")

    
text.replace(text[text.find("["):text.find("]")+1], " ").strip()
text.replace(text[text.find("("):text.find(")")+1], " ").strip()
text = re.sub(r'[0-9]+', '', text)
#text = re.sub("\!|\'|\?","",text)
specialChars = '/\[]()"",.:'
for specialChar in specialChars:
    text = text.replace(specialChar, "")
#text.replace("/[\[\].!'@,><|://\\;&*()_+=]/g", "")
#re.sub("[\(\[].*?[\)\]]", "", text)
#re.sub(' +', ' ', text)

#print(text)
```

```python
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)


pattern = [{"LEMMA": {"IN": ["love", "I"]}},
            {"POS": "NOUN"}]

matcher.add("Love", [pattern])
matches = matcher(doc)
print("Total matches found: ", len(matches))
#print(matches)

for match_id, start, end in matches:
    string_id = nlp.vocab.strings[match_id]  # Get string representation
    span = doc[start:end]  # The matched span
    print(match_id, string_id, start, end, span.text)
    #print("Match Found: ", doc[start:end].text)
```

```python
#This cell counts how often words appear in our doc

from collections import Counter

# all tokens that arent stop words or punctuations
words = [token.text
         for token in doc
         if not token.is_stop and not token.is_punct]

# noun tokens that arent stop words or punctuations
nouns = [token.text
         for token in doc
         if (not token.is_stop and
             not token.is_punct and
             token.pos_ == "NOUN")]

# five most common tokens
word_freq = Counter(words)
common_words = word_freq.most_common(5)

# five most common noun tokens
noun_freq = Counter(nouns)
common_nouns = noun_freq.most_common(5)
```

```python
import spacy.attrs

# Returns integers that map to parts of speech
counts_dict = doc.count_by(spacy.attrs.IDS['POS'])

# Print the human readable part of speech tags
for pos, count in counts_dict.items():
    human_readable_tag = doc.vocab[pos].text
    print(human_readable_tag, count)
```
