```python
import json
import os
import sys
import pathlib



json_lyrics_files = []    # names of files
json_lyrics_data = {}     # dictionary for data from files, key is filename

for r, d, f in os.walk('json_lyrics_texts'): # folder where all json files exist
        for file in f:
            if file.endswith('.json'):
                json_lyrics_files.append(os.path.join(r, file))
                print(json_lyrics_files)

for fn in json_lyrics_files:
    with open(fn, 'r') as f:
        data = json.load(f) 
        json_lyrics_data[fn] = data

# save new json file
with open('final_data.json', 'w') as f:
    json.dump(json_lyrics_data, f)
```
